```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_index = pd.read_csv('economic_index.csv')
df_index.head()
```

```python
#drop unneccessary columns
df_index.drop(columns=['Unnamed: 0', 'year', 'month'], inplace=True, axis=1)
```

```python
df_index.head()
```

```python
##check null values
df_index.isna().sum()
```

```python
## Lets do some viualization
import seaborn as sns
sns.pairplot(df_index)
```

```python
df_index.corr()
```

```python
## Visualiza the datapoints more closely

plt.scatter(df_index['interest_rate'], df_index['unemployment_rate'], color='b')
plt.xlabel('Interest rate')
plt.ylabel('Unemployment rate')
```

```python
##independent and dependent features

X = df_index.iloc[:, :-1]
y = df_index.iloc[:, -1]

X.head()
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
import seaborn as sns
sns.regplot(data=df_index, x=df_index['interest_rate'], y=df_index['index_price'])
```

```python
sns.regplot(data=df_index, x=df_index['interest_rate'],y=df_index['unemployment_rate'])
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.fit_transform(X_test)

X_train
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
## cross validation : 
from sklearn.model_selection import cross_val_score
valid_score = cross_val_score(lr, X_train, y_train, scoring='neg_mean_squared_error', cv=3)
valid_score
```

```python
np.mean(valid_score)
```

```python
## prediction
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
print(score)
#display adjusted R-squared
print(1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```
